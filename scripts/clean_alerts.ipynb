```python
# import 
import os 
import geopandas as gpd 
from random import random
import numpy as np
```

```python
# processing
# output file 
output_file = '../data/alerts.gpkg'
if not os.path.exists(output_file): 
    # alerts gdf
    alerts_gdf = gpd.read_file('../data/Data.zip')
    alerts_gdf = alerts_gdf.rename(columns={
        'count': 'id', 
        'REGION': 'region',
        'SUPERFICIE': 'superficie_ha',
        'PROBABILIT': 'probabilite'
    })
    alerts_gdf = alerts_gdf[['id', 'region', 'superficie_ha', 'probabilite', 'geometry']]
    display(alerts_gdf)

    # stats
    alerts_gdf['superficie_ha'] = alerts_gdf['superficie_ha']/10_000
    alerts_gdf['probabilite'] = np.random.uniform(0, 100, size=len(alerts_gdf))
    alerts_gdf['risk'] = np.where(alerts_gdf['probabilite'] > 70, 'High', 'Low')
    display(alerts_gdf)

    # keep cols
    keep_cols = ['id', 'region', 'superficie_ha', 'probabilite', 'risk', 'geometry']
    alerts_gdf = alerts_gdf[keep_cols]

    # save
    alerts_gdf.to_file(output_file)

else : 
    alerts_gdf = gpd.read_file(output_file)
    display(alerts_gdf)
```

```python
region_summary = alerts_gdf.groupby('region')['risk'].agg(
    total_sites='count',
    high_risk_count=lambda x: (x == 'High').sum()
).reset_index()

display(region_summary)
```
