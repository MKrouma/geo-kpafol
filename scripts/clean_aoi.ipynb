```python
# import 
import os 
import geopandas as gpd 
```

```python
# read aoi
aoi_path = '../data/aoi.gpkg'
aoi_gdf = gpd.read_file(aoi_path)

display(aoi_gdf)
```

```python
# keep cols
keep_cols = ['COUNTRY', 'NAME_1', 'NAME_2', 'geometry']
aoi_gdf = aoi_gdf[keep_cols]
aoi_gdf = aoi_gdf.rename(columns={
    'COUNTRY': 'Pays',
    'NAME_1': 'District',
    'NAME_2': 'Region'
})

display(aoi_gdf)
```

```python
# save
aoi_gdf.to_file('../data/aoi_clean.gpkg')
```
